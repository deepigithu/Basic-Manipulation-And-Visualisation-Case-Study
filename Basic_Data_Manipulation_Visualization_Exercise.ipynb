### 30. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
users['count_of_male']=np.where(users['gender']=='M',1,0)
users
```

```python
#pivot_table
users['count_of_male']=np.where(users['gender']=='M',1,0)

total_male_by_occupation=pd.pivot_table(data=users,index='occupation',values='count_of_male',aggfunc='sum')

total_user_by_occupation=pd.pivot_table(data=users,index='occupation',values='count_of_male',aggfunc=len)

total_men_ratio_per_occupation=total_male_by_occupation/total_user_by_occupation

total_men_ratio_per_occupation.sort_values(by='count_of_male',ascending=False)
```

```python
#groupby
users['count_of_male']=np.where(users['gender']=='M',1,0)
a=users.groupby('occupation')[['count_of_male']].sum('count_of_male')
b=users.groupby('occupation')[['count_of_male']].count()
c=(a/b).sort_values(by='count_of_male',ascending=False)
c
```

### 33.  For each occupation present the percentage of women and men

```python
#pivot_table
total_male_by_occupation=pd.pivot_table(data=users,index='occupation',values='count_of_male',aggfunc='sum')

total_user_by_occupation=pd.pivot_table(data=users,index='occupation',values='count_of_male',aggfunc=len)

total_men_ratio_per_occupation=total_male_by_occupation/total_user_by_occupation

total_men_ratio_per_occupation

```

```python
temp_df=total_men_ratio_per_occupation.copy()

temp_df['count_of_male']=1.0

percentage_of_women=temp_df-total_men_ratio_per_occupation

percentage_of_women=percentage_of_women.rename(columns={'count_of_male':'count_of_female'})

percentage_of_women
```

```python
#groupby
gender_occupation=users.groupby(['occupation','gender']).agg({'gender':'count'})
occupation_count=users.groupby(['occupation']).count()
occupation_gender=gender_occupation.div(occupation_count,level="occupation")
occupation_gender['gender']
```
